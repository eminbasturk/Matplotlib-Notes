# INTRODUCTION TO MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

**1)** <code>fig, ax = plt.subplots()</code> Boş grafik oluşturmaya yarayan fonksiyon. <code>fig</code> şeklin kendisini, <code>ax</code> eksenleri temsil eder. Üstüne grafik eklenebilir. Fonksiyon içine **satır** ve **sütun** bilgilerini ekleyerek bir matris içine grafikleri yerleştirebilirsin. Matris içindeki grafiklere erişim indexlemeler ile olur.
* <code>sharey=<b>True</b></code> Her iki grafiğinde aynı oranlama ile görünmesini sağlayan argüman.

**2)** <code>plt.show()</code> Çizilen grafiğin gösterilmesini sağlayan fonksiyon.

```python
fig, ax = plt.subplots(1, 2, sharey = True)
plt.show()
```

**3)** <code>ax.plot()</code> Eksenlere veriler ekleyip şekil çizen fonksiyon. <code>fig, ax = plt.subplots()</code> ile birlikte kullanılır. Eksenlere birden fazla veri eklenebilir.
* <code>marker</code> Grafikte veri noktalarına istediğin karakteri koymayı sağlayan argüman. Veri noktalarını belirtmek için kullanılır. Çizgiler sadece bağlayıcıdır.
* <code>linestyle</code> Grafikteki bağlantı çizgilerini değiştirmek için kullanılır. Kesik çizgili (<b>"--"</b>) ve nokta çizgili <b>"-."</b> gibi çizgiler kullanılabilir. Bağlantı çizgisi koymaya da bilirsin. (<b>"None"</b>)
* <code>color</code> Verilerin rengini değiştirmek için kullanılan argüman.

**4)** <code>ax.set_xlabel()</code> ve <code>ax.set_ylabel()</code> **x** ve **y** eksenlerine etiket atayan metot. Grafiğin daha okunabilir olması için önemlidir.

**5)** <code>ax.set_title ()</code> Başlık ekleyen metot. Grafiğin daha okunabilir olması için önemlidir.

### Elde edilen grafik fonksiyona eklenerek taslaklaştırılabilir.

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color = color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params("y", colors = color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, spacex.index, spacex["Time (UTC)"], "blue", "Date (year)", "Time (UTC)")
ax2 = ax.twinx()
plot_timeseries(ax2, spacex.index, spacex["Payload Mass (kg)"], "red", "Date (year)", "Payload Mass (kg)")
plt.show()
```

**9)** <code>.annotate()</code> Grafikteki belirli noktalara dikkat çekmek için konulan ek açıklamalar.
* <code>xy = [pd.Timestamp("x"), y]</code> Dikkat çekmek istenen nokta.
* <code>xytext = [pd.Timestamp("x"), y]</code> Noktaya konulan yazı okunmadığı durumda kullanılan konum.
* <code>arrowprops = {"arrowstyle":"->", "color":"gray"}</code> Geçici konumdan gerçek konumu işaret etmek için kullanılan ok ve özellikleri.

```python
fig, ax = plt.subplots()
plot_timeseries(ax, spacex.index, spacex["Time (UTC)"], "blue", "Date (year)", "Time (UTC)")
ax.annotate("> 14:00", xy = [pd.Timestamp("2017-09-07"), "14:00:00"], xytext = [pd.Timestamp("2014-07-14"), "1:00:00"], arrowprops = {"arrowstyle":"->", "color":"gray"})
ax2 = ax.twinx()
plot_timeseries(ax2, spacex.index, spacex["Payload Mass (kg)"], "red", "Date (year)", "Payload Mass (kg)")
plt.show()
```

```python
import pandas as pd
teams = pd.DataFrame({"Team": ["SUP", "AUR", "CRW", "HWA", "FB", "GAL", "DP", "GS"],
                      "Win": [10, 6, 5, 5, 4, 4, 2, 2],
                      "Draw": [2, 4, 5, 4, 6, 5, 6, 4],
                      "Lose": [2, 4, 4, 5, 4, 5, 6, 8]})
teams = teams.set_index("Team")
teams.head()
```

**10)** <code>.bar()</code> Sayısal değişkenin gözlemlere veya kategorik değişken sınıflarına göre göre değerlerini karşılaştıran grafik. Her bir çubuk bir gözlemi temsil eder. Çubukların yüksekliği gözlemin değişken değerine göre belli olur. İki veri setinden benzer değişkenleri aynı grafik üzerinde de karşılaştırabilirsin.
* <code>bottom</code> Grafikte birden fazla değişken kullanmaya olanak tanır. Önceki ifadedeki değişken veya değişkenler buraya yazılır. Argümanı kullanmak için fonksiyonu yeniden çağırmak gerekir.
* <code>label</code> Grafiklerin birbirleri ile karışmaması için grafiğe eklenen etiket. <code>.legend()</code> metotunun kullanılmasını gerektirir. 
* <code>yerr</code> Grafikte çizilen çubukta hata payını verir. **y** ekseninde <code>.mean()</code> varken genellikle <code>.std()</code> metodunu içerir.

**11)** <code>.set_xticklabels(labels, rotation)</code> **x** eksenindeki etiketlerin grafikteki görünümünü değiştirmek için kullanılır. Manuel olarak etiket verebilir ve etiketlere döndürme işlemi yapabilirsin.

```python
fig, ax = plt.subplots()
ax.bar(teams.index, teams["Win"], label = "Win")
ax.bar(teams.index, teams["Draw"], bottom = teams["Win"], label = "Draw")
ax.bar(teams.index, teams["Lose"], bottom = teams["Win"] + teams["Draw"], label = "Lose")
ax.set_xticklabels(teams.index, rotation = 90)
ax.set_ylabel("Number of wins")
ax.legend()
plt.show()
```

**19)** <code>.unique()</code> Veri setindeki kategorik değişken içindeki sınıflardan benzersiz olanları getiren metot.

```python
import pandas as pd
iris = pd.read_csv("iris.csv")
flowers = iris["Species"].unique()
flowers
```

**<code>.unique()</code> sayesinde otomatikleşme**

```python
# Her bir çiçek türünün çanak yaprak uzunluğunun ortalaması ve standart sapması alınır.
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
for flower in flowers:
    flower_df = iris[iris["Species"] == flower]
    ax.bar(flower, flower_df["SepalLengthCm"].mean(), yerr = flower_df["SepalLengthCm"].std())
ax.set_ylabel("Sepal length (cm)")
ax.set_xticklabels(flowers, rotation = 90)
plt.show()
```
